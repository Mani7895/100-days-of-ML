```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
df["mainroad"] = df["mainroad"].map({"yes":1, "no": 0})
df["guestroom"] = df["guestroom"].map({"yes":1, "no": 0})
df["basement"] = df["basement"].map({"yes":1, "no": 0})
df["hotwaterheating"] = df["hotwaterheating"].map({"yes":1, "no": 0})
df["airconditioning"] = df["airconditioning"].map({"yes":1, "no": 0})
df["prefarea"] = df["prefarea"].map({"yes":1, "no": 0})
```

```python
status = pd.get_dummies(df["furnishingstatus"], drop_first=True)
df = pd.concat([df, status], axis = 1)
df.drop(["furnishingstatus"], axis = 1, inplace = True)
df.head()
```

```python
df = (df - df.mean())/df.std()
df.head()
```

```python
X = df["area"]
y = df["price"]
```

```python
import seaborn as sns
sns.pairplot(df, x_vars = "area", y_vars= "price", size = 7, aspect =0.7, kind = "scatter")
```

```python
X = np.array(X)
y = np.array(y)
```

```python
def gradient(X, y, m_cur= 0, c_cur= 0, iters= 1000, lr= 0.01):
    N = float(len(y))
    gd_df = pd.DataFrame(columns= ["m_cur", "c_cur", "cost"])
    for i in range(iters):
        y_cur = (m_cur * X) + c_cur
        cost = sum([data**2 for data in (y-y_cur)])/N
        m_grad = -(2/N) * sum(X * (y-y_cur))
        c_grad = -(2/N) * sum(y - y_cur)
        m_cur = m_cur - (lr * m_grad)
        c_cur = c_cur - (lr * c_grad)
        gd_df.loc[i] = [m_cur, c_cur, cost]
    
    return gd_df

gradients = gradient(X, y)
gradients
```

```python
gradients.reset_index().plot.line(x = "index", y = ["cost"])
```

```python
X = df[["area", "bedrooms"]]
y = df["price"]
```

```python
X["intercept"] = 1
X = X.reindex_axis(["intercept", "area", "bedrooms"], axis= 1)
X.head()
```

```python
X = np.array(X)
y = np.array(y)
```

```python
theta = np.matrix(np.array([0, 0, 0]))
alpha = 0.01
iter = 1000
```

```python
def compute_cost(X, y, theta):
    return np.sum(np.square(np.matmul(X, theta) - y)) / (2*len(y))
```

```python
def gradDescent_multi(X, y, theta, alpha, iter):
    theta = np.zeros(X.shape[1])
    m = len(X)
    gdm_df = pd.DataFrame(columns=["Bets", "cost"])
    
    for i in range(iter):
        grad = (1/m) * np.matmul(X.T, np.matmul(X, theta)-y)
        theta = theta - alpha * grad
        cost = compute_cost(X, y, theta)
        gdm_df.loc[i] = [theta, cost]
        
    return gdm_df

gradDescent_multi(X, y, theta, alpha, iter)
```
